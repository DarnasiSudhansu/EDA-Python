```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cleaning the Data
HA['ratings']=HA['ratings'].fillna('0')
HA=HA.fillna('')
HA
```

```python
HA=HA.drop(['image','link','sub_category','main_category'],axis=1)
HA
```

```python
HA['ratings']=HA['ratings'].replace('Get','')
HA['no_of_ratings']=HA['no_of_ratings'].replace('FREE Delivery by Amazon','')
HA['no_of_ratings']=HA['no_of_ratings'].replace('Only 1 left in stock.','')
HA['no_of_ratings']=HA['no_of_ratings'].replace('Only 2 left in stock.','')
HA
```

```python
for x in HA.index:
    if HA.loc[x,'no_of_ratings']=='':
        HA.drop(x,inplace=True)

HA
```

```python
for x in HA.index:
    if HA.loc[x,'discount_price']=='':
        HA.drop(x,inplace=True)

HA
```

```python
for x in HA.index:
    if HA.loc[x,'actual_price']=='':
        HA.drop(x,inplace=True)

HA
```

```python
HA['actual_price']=HA['actual_price'].str.strip('₹,')
HA['discount_price']=HA['discount_price'].str.strip('₹,')
HA
```

```python
HA.isnull().sum()
```

```python
HA.dtypes
```

```python
#Conversion of Data Types
HA['no_of_ratings']=HA['no_of_ratings'].str.replace(',','').astype(float)
HA['actual_price']=HA['actual_price'].str.replace(',','').astype(float)
HA['discount_price']=HA['discount_price'].str.replace(',','').astype(float)
HA['ratings']=HA['ratings'].astype(float)
HA['name']=HA['name'].astype(str)
HA
```

```python
HA.dtypes
```
