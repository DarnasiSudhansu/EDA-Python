```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the Data
HD=pd.read_csv('Home Dcor.csv')
```

```python
#Studying the Data
HD.info()
```

```python
HD['ratings'].unique()
```

```python
HD['no_of_ratings'].unique()
```

```python
#Cleaning the Data
HD['ratings']=HD['ratings'].fillna('0')
HD=HD.fillna('')
HD
```

```python
HD=HD.drop(['image','link','sub_category','main_category'],axis=1)
HD
```

```python
HD['ratings']=HD['ratings'].replace('Get','')
HD['no_of_ratings']=HD['no_of_ratings'].replace('FREE Delivery by Amazon','')
HD['no_of_ratings']=HD['no_of_ratings'].replace('Only 1 left in stock.','')
HD
```

```python
for x in HD.index:
    if HD.loc[x,'no_of_ratings']=='':
        HD.drop(x,inplace=True)

HD
```

```python
for x in HD.index:
    if HD.loc[x,'discount_price']=='':
        HD.drop(x,inplace=True)

HD
```

```python
for x in HD.index:
    if HD.loc[x,'actual_price']=='':
        HD.drop(x,inplace=True)

HD
```

```python
HD['actual_price']=HD['actual_price'].str.strip('₹,')
HD['discount_price']=HD['discount_price'].str.strip('₹,')
HD
```

```python
HD.isnull().sum()
```

```python
HD.dtypes
```

```python
#Conversion of Data Types
HD['no_of_ratings']=HD['no_of_ratings'].str.replace(',','').astype(float)
HD['actual_price']=HD['actual_price'].str.replace(',','').astype(float)
HD['discount_price']=HD['discount_price'].str.replace(',','').astype(float)
HD['ratings']=HD['ratings'].astype(float)
HD['name']=HD['name'].astype(str)
HD
```

```python
HD.dtypes
```

```python
HD=HD.reset_index(drop=True)
HD
```
